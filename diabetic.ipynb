```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Read data ,split features on X and labels on y

```python
dd=pd.read_csv("diabetes.csv")
X=dd.drop(columns='Outcome',axis=1)
y=dd['Outcome']
```

Now standardize the data

```python
scalar=StandardScaler()              # or we can use standard_dd=scalar.fit_transform(X)
scalar.fit(X)
standard_dd=scalar.transform(X)
print(standard_dd)
X=standard_dd
```

train test split method

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y, random_state=40)
```

Traing the model


```python
classifier=svm.SVC(kernel='linear')
classifier.fit(X_train,y_train)
```

Accuracy Score

```python
X_train_prediction=classifier.predict(X_train)
train_training_accuracy=accuracy_score(X_train_prediction,y_train)
print("ACCURACY",train_training_accuracy)
```

```python
X_test_prediction=classifier.predict(X_test)
test_training_accuracy=accuracy_score(X_test_prediction,y_test)
print("ACCURACY",test_training_accuracy)
```

Making a predictive system

```python
input_data=(4, 110, 92, 0, 0, 37.6, 0.191, 30) 
setdata_array_usingnumpy=np.asarray(input_data) #given input is random data so we can set like a array 
reshape_data=setdata_array_usingnumpy.reshape(1,-1) #trained model use several data but now we give only one data so we need to reshape
std_data=scalar.transform(reshape_data) # convert the input into standard model so the value can be similar 
predicton=classifier.predict(std_data) # now predict
print(predicton)
```

```python
# User input for prediction
feature_names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
user_input = []
for feature in feature_names:
    value = float(input(f"Enter {feature}: "))
    user_input.append(value)

# Convert input to DataFrame with correct columns
test_df = pd.DataFrame([user_input], columns=dd.drop(columns='Outcome').columns)
# Standardize input
test_std = scalar.transform(test_df)
# Predict
test_pred = classifier.predict(test_std)

if test_pred[0] == 1:
    print("The patient is diabetic.")
else:
    print("The patient is not diabetic.")
```
